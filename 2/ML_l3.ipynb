#Задание
Построить графики зависимости:

1. Метрики качества классификации от количества деревьев
2. Качества обучения от глубины деревьев

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_wine()
data, labels = dataset['data'], dataset['target']
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, random_state=0)
```

#Задание №1
Построить график зависимости метрики качества классификации от количества деревьев

```python
scores = []
n_estimators = np.arange(1,30)
for i in n_estimators:
    clf = RandomForestClassifier(max_depth=10, random_state=0, n_estimators=i)
    clf.fit(X_train, Y_train)
    scores.append(clf.score(X_test,Y_test))
plt.plot(n_estimators, scores)
plt.xlabel('Количество деревьев')
plt.ylabel('Точность')
plt.title('Метрика качества от количества деревьев')
plt.show()
```

# Задание №2
Построить график зависимости качества обучения от глубины деревьев

```python
scores = []
n_depths = range(2, 30)
for i in n_depths:
    clf = RandomForestClassifier(max_depth = i, random_state=0)
    clf.fit(X_train, Y_train)
    scores.append(clf.score(X_test, Y_test))


plt.plot(n_depths, scores)
plt.xlabel('Максимальная глубина')
plt.ylabel('Точность')
plt.title('Качество обучения от глубины деревьев')
plt.show()
```
